#### <font color='maroon'><b>Concatenating the Amazon and Flipkart datasets :

```python
import pandas as pd
import numpy as np
```

```python
# amazon dataset

dfa = pd.read_csv('Amazon_microwave_oven.csv')
dfa.head()
```

```python
# flipkart dataset

dff = pd.read_csv('Flipkart_microwave_oven.csv')
dff.head()
```

```python
dff.rename(columns={'discounts in %':'discount in %'},inplace=True)
```

```python
dff.rename(columns={'Model Name':'Model'},inplace=True)
dff.rename(columns={'Type':'Oven Cooking Mode'},inplace=True)
```

```python
# concatenating the two dataframes

df = pd.concat([dff,dfa],axis=0).reset_index(drop=True)
df.head()
```

```python
df.rename(columns={'star_rating of product out of 5':'star_rating','sales_price in Rs/-':'sales_price','mrp in Rs/-':'mrp','discount in %':'discount_percent'},inplace=True)
```

```python
df.rename(columns={'Model':'model_name'},inplace=True)
```

#### <font color='maroon'>Selecting the columns needed for data analysis :

```python
df_sel
```

```python
# check for duplicates

df_sel.duplicated(subset='product_title').sum()
```

```python
df_sel['discount_percent'] = df_sel['discount_percent'].astype(float).astype('Int64')
```

```python
df_sel['number_of_ratings'] = df_sel['number_of_ratings'].astype(float).astype('Int64')
```

```python
df_sel['number_of_reviews'] = df_sel['number_of_reviews'].astype(float).astype('Int64')
```
