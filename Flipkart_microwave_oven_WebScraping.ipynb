#### <font color='maroon'><b>Flipkart - Microwave oven - Web Scraping - using Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import re
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
# Configure Chrome options to disable images and ads
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--blink-settings=imagesEnabled=false")
chrome_options.add_argument("--dns-prefetch-disable")
```

```python
# Create a new Chrome driver instance using the managed ChromeDriver
driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
```

```python
product_urls = []
product_title = []
ratings = []
no_of_ratings = []
sales_price = []
mrp = []
discounts = []
star_rating = []
seller_name = []
specifications = []
```

```python
def scrape_flipkart_product_details(driver):
    
    product_links = driver.find_elements(By.CSS_SELECTOR, 'a._1fQZEK')
    product_url = [link.get_attribute('href') for link in product_links if link.get_attribute('href')]
    product_urls.extend(product_url)
    
    for url in product_url:
        driver.get(url)
        driver.implicitly_wait(10)
        
        
    # titles
        try:
            title = driver.find_element(By.CSS_SELECTOR, "span.B_NuCI").text
            product_title.append(title)
        except:
            product_title.append('NA')
            
    
    # sales price

        try:
            sl_pr = driver.find_element(By.CSS_SELECTOR, "div._30jeq3._16Jk6d").text
            sales_price.append(sl_pr)
        except:
            sales_price.append('NA')
            
    
    # market price
    
        try:
            mk_pr = driver.find_element(By.CSS_SELECTOR, "div._3I9_wc._2p6lqe").text
            mrp.append(mk_pr)
        except:
            mrp.append('NA')
            
    # discounts
    
        try:
            discount = driver.find_element(By.CSS_SELECTOR, "div._3Ay6Sb._31Dcoz span").text
            discounts.append(discount)
        except:
            discounts.append('NA')
    
    
    # product ratings
    
        try:
            rating = driver.find_element(By.CSS_SELECTOR, "div._3LWZlK").text
            ratings.append(rating)
        except:
            ratings.append('NA')
            
            
    # number of ratings and reviews
    
        try:
            no = driver.find_element(By.CSS_SELECTOR, "span._2_R_DZ span").text
            no_of_ratings.append(no)
        except:
            no_of_ratings.append('NA')
    
    
    # seller raings
   
        try:
            star = driver.find_element(By.CSS_SELECTOR, "div._3LWZlK._1D-8OL").text
            star_rating.append(star)
        except:
            star_rating.append('NA')
            
    # seller
    
        try:
            sell = driver.find_element(By.CSS_SELECTOR, "div#sellerName._1RLviY span span").text
            seller_name.append(sell)
        except:
            seller_name.append('NA')
            
    # specifications
    
        try:
            driver.implicitly_wait(10)
            read_more_button = driver.find_element(By.CSS_SELECTOR, 'button._2KpZ6l._1FH0tX')
            read_more_button.click()
            driver.implicitly_wait(20)
            general = driver.find_element(By.CSS_SELECTOR, "div._1UhVsV")
            rows = general.find_elements(By.CSS_SELECTOR, "tr._1s_Smc.row")

            details = {}
            for row in rows:
                key_element = row.find_element(By.CSS_SELECTOR, 'td._1hKmbr')
                value_element = row.find_element(By.CSS_SELECTOR, 'td.URwL2w')
                key = key_element.text.strip()
                value = value_element.text.strip()
                details[key] = value

            specifications.append(details)
        except:
            specifications.append({})
            
    return product_urls, product_title, sales_price, mrp, ratings, no_of_ratings, discounts, star_rating, seller_name, specifications
```

```python
for i in range(1,9):
    driver.get(f"https://www.flipkart.com/microwave-ovens/pr?sid=j9e,m38,o49&page={i}")
    driver.implicitly_wait(10)
    product_urls, product_title, sales_price, mrp, ratings, no_of_ratings, discounts, star_rating, seller_name, specifications = scrape_flipkart_product_details(driver)
```

#### <font color='maroon'><b>Organising the data and Data Cleaning :

```python
df1 = pd.DataFrame({'product_url':product_urls,
                   'product_title':product_title,
                   'star_rating of product out of 5':ratings,
                   'no_of_ratings and reviews':no_of_ratings,
                   'sales_price in Rs/-':sales_price,
                   'mrp in Rs/-':mrp,
                   'discounts in %':discounts,
                   'seller_name':seller_name,
                   'seller_rating':star_rating,
                   'product_specifications':specifications})
```

```python
df1.head()
```

```python
# Fill NA values with an empty string
df1['no_of_ratings and reviews'].fillna('', inplace=True)

# Separate ratings and reviews using str.extract()
df1[['Ratings', 'Reviews']] = df1['no_of_ratings and reviews'].str.extract(r'(\d+(?:,\d+)?) Ratings & (\d+(?:,\d+)?) Reviews')

# Remove comma separators and convert to numeric values
df1['Ratings'] = df1['Ratings'].str.replace(',', '').astype(float)
df1['Reviews'] = df1['Reviews'].str.replace(',', '').astype(float)
```

```python
# Cleaning the data
df1['sales_price in Rs/-'] = df1['sales_price in Rs/-'].str.replace('₹', '').str.replace(',', '')
df1['mrp in Rs/-'] = df1['mrp in Rs/-'].str.replace('₹', '').str.replace(',', '')
df1['discounts in %'] = df1['discounts in %'].str.replace('% off', '')
df1['Ratings'] = df1['Ratings'].astype(str).str.replace('\.0$', '')
df1['Reviews'] = df1['Reviews'].astype(str).str.replace('\.0$', '')
```

```python
# Renaming
df1.rename(columns={'Ratings':'number_of_ratings','Reviews':'number_of_reviews'},inplace=True)
```

```python
# Dropping the reduntant column
df1.drop(['no_of_ratings and reviews'],axis=1,inplace=True)
```

```python
# Converting product specifications to dataframe
df_spec = pd.DataFrame(specifications)
df_spec.head()
```

```python
# extracting only number for the capacity column
df_spec['Capacity'] = df_spec['Capacity'].apply(lambda x:re.findall('\d+\.?\d*',x)[0])
```

```python
df_spec.rename(columns={'Capacity':'Capacity in litres'},inplace=True)
```

```python
pd.set_option('display.max_columns',None)
```

```python
# Final Dataframe

df = df1.join(df_spec)
df.head()
```
